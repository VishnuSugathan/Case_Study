# ● Check whether all the zones generate the same amount of sales.

##### H0(Null Hypothesis) = All the zones generate the same amount of sales.

##### Ha(Alternate Hypothesis) = All the zones generate different amount of sales.

##### Let the significant level be 0.05(5%)

```python
from scipy.stats import f_oneway
A = np.array(data['Zone - A'])
B = np.array(data['Zone - B'])
C = np.array(data['Zone - C'])
D = np.array(data['Zone - D'])
st,p = f_oneway(A,B,C,D)
print("Statistic value = ",st)
print("P value = ",p)
if(p < 0.05):
    print("P value is less than 0.05")
    print("Null Hpothesis is rejected.")
    print("All the zones generate different amount of sales ")
else:
    print("P value is greater than 0.05")
    print("Null Hpothesis is accepted.")
    print("All the zones generate the same amount of sales. ")
```
